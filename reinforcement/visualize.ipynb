```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Read the CSV file into a pandas dataframe
df = pd.read_csv('Q_learning_game_statistics.csv')

# Group the scores by every 100 games and compute the mean score for each group
grouped_scores = df.groupby(df.index // 100)['score'].mean()

fig = plt.figure(figsize=(20, 10))
# Create a line plot of the mean scores
plt.plot(grouped_scores)

# Set the plot title and axis labels

plt.title('Mean Scores per 100 Games during Training',fontsize=20)
plt.xlabel('Training Iteration (in hundreds)',fontsize=20)
plt.ylabel('Mean Score',fontsize=20)

plt.tick_params(axis='both', labelsize=18)

# Display the plot
plt.show()
```
